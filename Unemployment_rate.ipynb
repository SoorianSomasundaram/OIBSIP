### **Correlation Analysis Between Participation rate and Estimated Unemployment rate**




```python
scaler = StandardScaler()

df_std_encoded = df[[' Estimated Labour Participation Rate (%)', ' Estimated Unemployment Rate (%)', ' Estimated Employed']]

scaler.fit_transform(df_std_encoded)
```

```python
P_df = ' Estimated Labour Participation Rate (%)'
U_df = ' Estimated Unemployment Rate (%)'

corr_df = df_std_encoded[[P_df] + [U_df]].corr()

Participate_Unemploy_corr = corr_df.loc[P_df,U_df]
Participate_Unemploy_corr
```

```python
heatmap = go.Heatmap(x=corr_df.columns,
                     y=corr_df.columns,
                     z=corr_df.values,
                     colorscale='purples')

layout = go.Layout(title='correlation between unemployment and participation rate')

fig = go.Figure(data=[heatmap],layout=layout)

fig.show()
```

### **Correlation between unemployment rate and estimated employment**

#### Standardizing of column estimated employment

```python
E_df = ' Estimated Employed'
U_df = ' Estimated Unemployment Rate (%)'

corr_df = df_std_encoded[[E_df] + [U_df]].corr()

corr_df.loc[E_df,U_df]
```
